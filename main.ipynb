```python
# TODO complete this model
class MainModel(nn.Module):
    def __init__(self, net_G=None, lr_G=2e-4, lr_D=2e-4,
                 beta1=0.5, beta2=0.999, lambda_L1=100.):
        super().__init__()

        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.lambda_L1 = lambda_L1

        if net_G is None:
            self.net_G = init_model(Unet(input_c=1, output_c=2, n_down=8, num_filters=64), self.device)
        else:
            self.net_G = net_G.to(self.device)
        self.net_D = init_model(PatchDiscriminator(input_c=3, n_down=3, num_filters=64), self.device)
        # DÃ©finir les critÃ¨res pour le gÃ©nÃ©rateur et le discriminant
        self.GANcriterion = nn.BCELoss()
        self.L1criterion = nn.L1Loss()
        self.opt_G = optim.Adam(self.net_G.parameters(), lr=lr_G, betas=(beta1, beta2))
        self.opt_D = optim.Adam(self.net_D.parameters(), lr=lr_D, betas=(beta1, beta2))

    def set_requires_grad(self, model, requires_grad=True):
        for p in model.parameters():
            p.requires_grad = requires_grad

    def setup_input(self, data):
        self.L = data['L'].to(self.device)
        self.ab = data['ab'].to(self.device)

    def forward(self):
        self.fake_color = self.net_G(self.L)

    def backward_D(self):
        # TODO: define the backward of the descriminator
        # Effacer les gradients du modÃ¨le discriminant
        self.opt_D.zero_grad()

        # Calculer les prÃ©dictions du discriminant sur les vraies et fausses images
        real_pred = self.net_D(torch.cat([self.L, self.ab], dim=1))
        fake_pred = self.net_D(torch.cat([self.L, self.fake_color.detach()], dim=1))  # detach pour geler les gradients du gÃ©nÃ©rateur

        # Calculer les Ã©tiquettes pour la perte adversaire
        real_label = torch.ones_like(real_pred)  # Ã‰tiquette rÃ©elle pour les vraies images
        fake_label = torch.zeros_like(fake_pred)  # Ã‰tiquette fausse pour les fausses images

        # Calculer la perte adversaire pour le modÃ¨le discriminant
        loss_D_real = self.GANcriterion(real_pred, real_label)
        loss_D_fake = self.GANcriterion(fake_pred, fake_label)
        loss_D = (loss_D_real + loss_D_fake) / 2  # Perte moyenne

        # RÃ©tropropagation et mise Ã  jour des paramÃ¨tres du modÃ¨le discriminant
        loss_D.backward()
        self.opt_D.step()

        return loss_D

    def backward_G(self):
        # TODO: define the backward of the generator

        # Effacer les gradients du gÃ©nÃ©rateur
        self.opt_G.zero_grad()

        # Calculer la perte GAN adversaire du gÃ©nÃ©rateur
        fake_pred = self.net_D(torch.cat([self.L, self.fake_color], dim=1))
        real_label = torch.ones_like(fake_pred)  # Ã‰tiquette rÃ©elle pour les fausses images
        loss_G_GAN = self.GANcriterion(fake_pred, real_label)  # Utiliser les vraies Ã©tiquettes pour tromper le discriminant

        # Calculer la perte L1 entre les couleurs gÃ©nÃ©rÃ©es et les couleurs rÃ©elles
        loss_G_L1 = self.L1criterion(self.fake_color, self.ab)

        # Calculer la perte totale du gÃ©nÃ©rateur
        loss_G = loss_G_GAN + self.lambda_L1 * loss_G_L1

        # RÃ©tropropagation et mise Ã  jour des paramÃ¨tres du gÃ©nÃ©rateur
        loss_G.backward()
        self.opt_G.step()

        return loss_G

    def optimize(self):
        self.forward()
        self.net_D.train()
        self.set_requires_grad(self.net_D, True)
        self.opt_D.zero_grad()
        self.backward_D()
        self.opt_D.step()

        self.net_G.train()
        self.set_requires_grad(self.net_D, False)
        self.opt_G.zero_grad()
        self.backward_G()
        self.opt_G.step()
```
